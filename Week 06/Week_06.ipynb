# <b>TIC TAC TOE:</b><br>

```python
import numpy as np

playing_board = np.array([['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']])

player_1 = 'X'
player_2 = 'O'

def place(symbol):
    print(np.matrix(playing_board))
    while(1):
        row = int(input("\nEnter Your Row(1/2/3): "))   
        Column = int(input("Enter Your Column(1/2/3): "))
        
        if row>0 and row<4 and Column>0 and Column<4 and playing_board[row-1][Column-1]=='-':
            break
        else:
            print("Invalid Input!")
              
    playing_board[row-1][Column-1] = symbol
    


def check_rows(symbol):
    for each_row in range(3):
        count = 0
        for each_column in range(3):
            if playing_board[each_row][each_column] == symbol:
                count += 1
        
        if count == 3:
            print(symbol, "Won!!")
            return True
    
    return False


def check_columns(symbol):
    for each_column in range(3):
        count = 0
        for each_row in range(3):
            if playing_board[each_row][each_column] == symbol:
                count += 1
        
        if count == 3:
            print(symbol, "Won!!")
            return True
    
    return False
               

def check_diagonals(symbol):
    if playing_board[0][2]==playing_board[1][1] and playing_board[1][1]==playing_board[2][0] and playing_board[1][1]==symbol:
        print(symbol,"Won!!")
        return True
    
    if playing_board[0][0]==playing_board[1][1] and playing_board[1][1]==playing_board[2][2] and playing_board[1][1]==symbol:
        print(symbol,"Won!!")
        return True
    
    return False



def won(symbol):
    return check_rows(symbol) or check_columns(symbol) or check_diagonals(symbol)

         
def play():
    for turn in range(9):
        if(turn%2==0):
            print("\nX's turn: ")
            place(player_1)
            if won(player_1):
                break
        else:
            print("\nO's turn: ")
            place(player_2)
            if won(player_2):
                break
        
        if not won(player_1) and not won(player_2):
            print("Draw !!")
            
    
play()
```

### <b> 1. Factorial Using Iteration: </b><br>

```python
def factorial(num):    
    fact = 1
    for i in range(1, num+1):
        fact *= i
    return fact



num = int(input("Enter the number: "))
if(num<0):
        print("Invalid input!!")
else:
    print("The factorial of", num, "is: ", factorial(num))
    print("\nThe factorial {} of is: {}".format(num, factorial(num)))
```

<br><br>
### <b> 2. Factorial Using Recursion: </b><br>

```python
def factorial(num): 
    if(num == 0 or num == 1):
        return 1
    return num*factorial(num-1)

num = int(input("Enter the number: "))
result = factorial(num)
if(num<0):
        print("Invalid input!!")
else:
    print("The factorial of", num, "is: ", result)
```

# <b> BINARY SEARCH: </b><br>

```python
def binary_Search(arr, target):
    start = 0
    end = len(arr) - 1
    flag = count = 0
    
    while(start<=end):
        count += 1
        
        mid = (start+end)//2
        
        if(arr[mid] == target):
            flag = 1
            print("The target element is present at index: ", str(mid))
            print("Total number of iteration required to search the target elemet is: ", count)
            return
        
        elif(arr[mid]<target):
            start = mid + 1
        
        else:
            end = mid - 1
       
     
    print("The target element is not present.")
        

arr = []
for i in range(1, 101):
    arr.append(i)

binary_Search(arr, 100)
```

```python
import string

def shift(word, value):
    letters = string.ascii_lowercase
    new = ''
    for i in range(len(word)):
        if word[i] in letters:
            index = letters.index(word[i])
            new = new + letters[(index+value)%26]
        
        else:
            new = new + word[i]
        
    return new

str = "aaaaaabbbbbbccccccdddddd"
value  = 5
print(shift(str, value))
            
```
