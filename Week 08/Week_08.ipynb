<h1 align="center"><b><u> Lottery Simulation - Profit or Loss </u></b></h1>

---
---


```python
# A small prototype of the above stated problem:
# ₹100 is entry fee for each game
# ₹900 is winning money on correct bet

import random

user_bet = int(input("Your bet{1-10}: "))
lucky_draw = random.randint(1, 10)

game_money = 0
print("You Bet on: ", user_bet)
print("Lucky Draw is: ", lucky_draw)

if user_bet == lucky_draw:
    game_money = game_money + 900 - 100

else:
    game_money = game_money - 100

print(game_money)
```

```python
# Actual Working solution for a week:

import random

game_money = 0
week = 7

for day in range(week):
    user_bet = random.randint(1, 10)
    lucky_draw = random.randint(1, 10)

    print("Bet is on: ", user_bet)
    print("Lucky Draw is: ", lucky_draw)

    if user_bet == lucky_draw:
        game_money = game_money + 900 - 100

    else:
        game_money = game_money - 100

    print("Account Money in game: ₹{}".format(game_money))
    print()
```

```python
# Actual Working solution for a year:

import random
import matplotlib.pyplot as plt

# Axes X and Y
x = []
y = []

game_money = 0
month = 30
year = 12

for month in range(year+1):
    x.append(month+1)
    for day in range(month+1):
        user_bet = random.randint(1, 10)
        lucky_draw = random.randint(1, 10)

        if user_bet == lucky_draw:
            game_money = game_money + 900 - 100

        else:
            game_money = game_money - 100
    y.append(game_money)

    print("Account Money in {} month in game: ₹{}".format(month+1, game_money))
    print()


plt.plot(x, y)
plt.show()
```

<h3 align="center"><b><u> Flipping The Image </u></b></h3>

```python
# Flipping the Image:
from PIL import Image

# Opeing the image:
img = Image.open("evidence.jpg")

# transposing the image:
transposed_img = img.transpose(Image.FLIP_LEFT_RIGHT)

# Converting into human understable form:
transposed_img.save('corrected_evidence.jpg')

# Checking that everything went right:
print("Done Flipping.")
```

<br>

<h3 align="center"><b><u> Enhancing the Image </u></b></h3>

```python
# IMAGE ENHACEMENT TECHNIQUE: Adaptive Histogram Equalisation( CLAHE- Contrast-limited adaptive histogram equalization)

import cv2

# Read the image
img = cv2.imread('bullet_shots.jpeg')

# Preparation for CLAHE( works well for black and white images)
cv2.createCLAHE()

# Converting into Gray scale image:
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Create a CLAHE object:
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

# Applying Enhacement:
enh_img = clahe.apply(gray_img)

# saving the image:
cv2.imwrite('enhanced_bullet.jpeg', enh_img)

# Confirmting
print("Done Enhancing.")
```

<h1 align="center"><b><u> FACEBOOK SENTIMENT ANALYSIS </u></b></h1>

---
---


```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.downloader.download('vader_lexicon')
file = '<fileName>.xslsx'       # Keep your excel file name here

# Reading the Excel File:
xl = pd.ExcelFile(file)

# Converting Excel files to dataframes
# Enter the column in which all the information is stored.
dfs = xl.parse(xl.sheet_names[0])

# Updating the dataframes by removing the blank rows:
# Timeline - name of the column
dfs = list(dfs['Timeline'])

print(dfs)

# Doing the sentiment analysis:
sid = SentimentIntensityAnalyzer()

str1 = "UTC+05:30"
for data in dfs:
    a = data.find(str1)

    # Working on rows that does not have UTC+05:30 and leaving all the other rows
    if (a == -1):
        ss = sid.polarity_scores(data)
        print(data)

        # Now take k to denote sentiment and ss[k] is the intensity of the sentiment
        for k in ss:
            print(k, ss[k])
```

```python

def replaceV(S):
    vowels = list("aeiouAEIOU")
    L = list(S)

    for i in range(len(L)-2):
        if L[i] in vowels and L[i+1] in vowels and L[i+2] in vowels:
            L[i] = '*'
            L[1+i] = "*"
            L[i+2] = "*"
            i += 2

    return ("".join(L).replace("***", "_"))


print(replaceV(input()), end="")


S = input()
print(replaceV(S))
```

<h1 align="center"><b><u> PROGRAMMING ASSIGNMENT - 03: </u></b></h1>
<br><br>


```python
def shiftZeroesRight(L):
    num_zeros = 0

    # Move non-zero elements to the left
    for i in range(len(L)):
        if L[i] != 0:
            L[i - num_zeros] = L[i]
        else:
            num_zeros += 1

    # Fill the end of the list with zeros
    for i in range(len(L) - num_zeros, len(L)):
        L[i] = 0

    return L


n = 10
new_List = list(map(int, input().split()))
result = shiftZeroesRight(new_List)
print(result, end="")
```
