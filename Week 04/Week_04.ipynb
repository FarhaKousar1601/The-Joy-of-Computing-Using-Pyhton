```python
def magic_square(num):
    
    square_matrix = []
    
    for i in range(num):
        l = []
        for j in range(num):
            l.append(0)
        square_matrix.append(l)

    i = num//2
    j = num - 1

    size = num * num
    count = 1
    
    while(count <= size):
        # Condition Nummber - 4:
        if(i == -1 and j == num):
            j = num - 2
            i = 0
        else:
            # When Column value exceeds:
            if(j == num):
                j = 0
            
            # When Row Becomes -1:
            if(i < 0):
                i = num - 1
            
        if(square_matrix[i][j] != 0):
            j = j - 2
            i = i + 1
            continue
        
        else:
            square_matrix[i][j] = count
            count = count + 1
            
        # Condition Number - 1:
        i = i - 1
        j = j + 1
    
    
    print("The sum of each Row/Column/Diagonal of", num, "X", num, "matrix is:", (num*((num**2)+1)//2)) 
    print()
    print("The Magic Square of", num, "X", num, "is:\n")      
    for i in range(num):
        for j in range(num):
            print(square_matrix[i][j], end=" ")
        print()


magic_square(int(input("Enter the square matrix size: ")))
```

### Leap Year Checker

```python
import random

year = random.randint(2001, 2023)

if(year%4==0 and year%100!=0 or year%400==0):
    print(year,"is a leap year.")
else:
    print(year,"is not a leap year.")
```

```python
import random
import datetime

birthday = []

i = 0
while(i<50):
    year = random.randint(1902, 2023)
    
    if(year%4==0 and year%100!=0 or year%400==0):
        leap = 1
    else:
        leap = 0
    
    month = random.randint(1, 12)
    
    # For leap year february:
    if(month == 2 and leap == 1):
        day = random.randint(1, 29)
    
    # For non-leap year february:
    elif(month == 2 and leap == 0):
        day = random.randint(1, 28)
    
    # For 7th and 8th month:
    elif(month==7 or month==8):
        day = random.randint(1, 31)
    
    #
    elif(month%2!=0 and month<7):
        day = random.randint(1, 31)
    
    #
    elif(month%2==0 and month>7 and month<12):
        day = random.randint(1, 31)
    
    #
    else:
        day = random.randint(1, 30)
        
    dd = datetime.date(year, month, day)
    day_of_year = dd.timetuple().tm_yday
    
    i = i+1
    birthday.append(day_of_year)

birthday.sort()

i = 0
while(i<50):
    print(birthday[i])
    i = i+1
```

# <b>Guess the Movie Name:</b><br>

```python
import random

movies = ["anand", "drishyam", "nayak", "anbe sivam", "gol maal", "vikram vedha", "black friday", "dangal", "taare zameen par"]

def create_que(movie):
    movie_size = len(movie)
    letters = list(movie)
    temp = []
    for i in range(movie_size):
        if(letters[i]==" "):
            temp.append(" ")
        else:
            temp.append("*")
            
    que = "".join(str(x) for x in temp)
    return que


def is_present(letter, movie):
    char_num = movie.count(letter)
    if(char_num==0):
        return False
    else:
        return True
    
    
def unlock(que, movie, letter):
    movie_ref = list(movie)
    que_list = list(que)
    temp = []
    movie_char_size = len(movie)
    for i in range(movie_char_size):
        if(movie_ref[i] == " " or movie_ref[i] == letter):
            temp.append(movie_ref[i])
        else:
            if(que_list[i]=="*"):
                temp.append("*")
            else:
                temp.append(movie_ref[i])
    
    new_que = "".join(str(x) for x in temp)
    return new_que
    

def play():
    p1name = input("Player-1, Enter your name: ")
    p2name = input("Player-2, Enter your name: ")
    print()
    
    pp1 = 0
    pp2 = 0
    
    turn = 0
    
    game = True
    while(game):
        # PLAYER - 1:
        if(turn%2==0):
            print(p1name,"Your turn")
            picked_movie = random.choice(movies)
            que = create_que(picked_movie)
            print(que)
            change_que = que
            
            choice = True
            while(choice):
                letter = input("Your Guess: ")
                if(is_present(letter, picked_movie)):
                    change_que = unlock(change_que, picked_movie, letter)
                    print(change_que)
                    decision = input("Press 1 to quess the movie or 2 for unlocking another character: ")
                    if(decision==1):
                        answer = input("Your Answer: ")
                        if(answer==picked_movie):
                            pp1 = pp1 + 1
                            print("Correct Answer !!")
                            choice = False
                            print(p1name, "Your score is: ",pp1)
                        
                        else:
                            print("Wrong Answer. Try Again...")
                            
                else:
                    print("Your Guessed letter", letter, "is not present.")
            
            move = input("Press 1 to continue or 0 to Quit: ")
            if(move==0):
                print(p1name, "Your score is: ", pp1)
                print(p2name, "Your score is: ", pp2)
                print("Thanks for playing the game!!!")
                game = False  
        
        
        # PLAYER - 2:        
        else:
            print(p2name,"Your turn")
            picked_movie = random.choice(movies)
            que = create_que(picked_movie)
            print(que)
            change_que = que
            
            choice = True
            while(choice):
                letter = input("Your Guess: ")
                if(is_present(letter, picked_movie)):
                    change_que = unlock(change_que, picked_movie, letter)
                    print(change_que)
                    decision = input("Press 1 to quess the movie or 2 for unlocking another character: ")
                    if(decision==1):
                        answer = input("Your Answer: ")
                        if(answer==picked_movie):
                            pp2 = pp2 + 1
                            print("Correct Answer !!")
                            choice = False
                            print(p2name, "Your score is: ",pp2)
                        
                        else:
                            print("Wrong Answer. Try Again...")
                            
                else:
                    print("Your Guessed letter", letter, "is not present.")
            
            move = input("Press 1 to continue or 0 to Quit: ")
            if(move==0):
                print(p1name, "Your score is: ", pp1)
                print(p2name, "Your score is: ", pp2)
                print("Thanks for playing the game!!!")
                game = False
                    
        turn = turn + 1
            

play()
```
